```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#%matplotlib inline
```

```python
x = np.arange(0, 10)
y = x * x
```

```python
plt.scatter(x, y, c='r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("Demo Plot")
plt.show()
```

```python

z = 2*x+3
plt.plot(x, y, 'rx--')
plt.plot(x, z, 'b')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'rx--')
plt.subplot(2, 2, 2)
plt.plot(x, z, 'b')
plt.show()
```

```python
x1 = ['A', 'B', 'C']
x2 = ['D', 'E', 'F']
y = [3, 4, 10]
z = [5, 4, 11]

plt.bar(x1, y, color='g', align='center')
plt.bar(x2, z, color='r')
plt.show()
```
